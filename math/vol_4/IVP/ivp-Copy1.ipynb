# 4.12

```python
def midpoint(A ,x0, t0, tf, h):
    xs = np.zeros((int((tf-t0)/h)+1, len(x0)))
    xs[0] = x0
    
    # First step is forward euler
    xs[1,:] = xs[0,:] + h * A @ xs[0]
    
    # Midpoint method
    for i in range(2, len(xs)):
        xs[i,:] = xs[i-2,:] + 2 * h * A @ xs[i-1,:]
        
    return xs
```

# 4.13

```python
A = np.array([[0, 3], [3, 0]])
x0 = np.array([1, 1])

x = midpoint(A, x0, 0, 3, .01)

plt.plot(x[:,0], x[:,1])
plt.show()
```

```python
A = np.array([[0, 3], [-3, 0]])
x0 = np.array([1, 1])

x = midpoint(A, x0, 0, 3, .01)

plt.plot(x[:,0], x[:,1])
plt.show()
```

# 4.14

```python
def trapezoidal(A ,x0, t0, tf, h):
    xs = np.zeros((int((tf-t0)/h)+1, len(x0)))
    xs[0] = x0
    
    # First step is forward euler
    xs[1,:] = xs[0,:] + h * A @ xs[0]
    
    # Trapezoidal method
    for i in range(2, len(xs)):
        xs[i,:] = np.linalg.inv(np.eye(2) - h/2*A) @ (xs[i-1,:] + h/2 * A @ xs[i-1,:])
        
    return xs
```

```python
A = np.array([[0, 3], [3, 0]])
x0 = np.array([1, 1])

x = midpoint(A, x0, 0, 3, .01)

plt.plot(x[:,0], x[:,1])
plt.show()
```

```python
A = np.array([[0, 3], [-3, 0]])
x0 = np.array([1, 1])

x = trapezoidal(A, x0, 0, 3, .01)

plt.plot(x[:,0], x[:,1])
plt.show()
```
