```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    df = pd.read_csv('crime_data.csv')
    df = df.set_index('Year')
    
    # Plot histogram of 
    df["crime_rate"] = df["Total"] / df["Population"]
    plt.suptitle("Crime Rates")
    df["crime_rate"].hist(grid=False)
    
    # Plot all crimes overall
    df.drop(["Population", "Total"], axis=1).plot(linewidth=1,title="Crimes")
    
    # Plot 4 crimes that are correlated
    df.plot(y=['Property', 'Larceny', 'Burglary', 'Violent'], subplots=True, 
            layout=(2,2), sharey=True, title="4 Most Common Crimes")
    
    
    
```

```python
prob1()
```

# Problem 2

```python
 def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    df = pd.read_csv('crime_data.csv')
    df = df.set_index('Year')
    df.plot(kind="scatter", x="Forcible Rape", y="Violent", 
            title="Correlation of Forcible Rape and Violent Crime")
    df.plot(kind="scatter", x="Forcible Rape", y="Burglary", 
            title="Correlation of Forcible Rape and Burglary")
    df.plot(kind="scatter", x="Forcible Rape", y="Aggravated Assault", 
            title="Correlation of Forcible Rape and Aggravated Assault")
    return (True, False, True)
```

```python
prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    df = pd.read_csv('crime_data.csv')
    df = df.set_index('Year')
    df[['Burglary', 'Violent', 'Vehicle Theft']].plot(kind='box')
    df.plot(kind='hexbin', x='Vehicle Theft', y='Robbery', gridsize=15, sharex=False)
    
    
    
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    e.g. what is the distribution of graduation rates or do schools 
    with lower student to faculty ratios have higher tuition costs.
    
    These plots should be easy to understand, 
    have clear variable names, and citations.
    """
    df = pd.read_csv("college.csv")
    print(df.columns)
    df["Total_cost"] = df["Outstate"] + df["Room.Board"] + df["Books"] + df["Personal"]
    
    
    # Plot histogram of 
    df["cost_rate"] = df["Expend"] / df["Outstate"]
    plt.suptitle("Histogram of rate of Instructional Expenditure/Tuition Cost")
    df["cost_rate"].hist(grid=False)
    plt.show()
    
    print("Alumni donations relationship with Graduation Rate among Private Schools")
    private_df = df[df['Private'] == "Yes"]
    private_df.plot(kind='hexbin', x="perc.alumni", y='Grad.Rate', gridsize=15, sharex=False)
    plt.show()
    
    plt.suptitle("Application vs. Acceptance")
    plt.xlabel("# of Applications")
    plt.ylabel("# of Accepted")
    plt.scatter(df["Apps"], df["Accept"], alpha=0.5)
    plt.show()
    
    
```
